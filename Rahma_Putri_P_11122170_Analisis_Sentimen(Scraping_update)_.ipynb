#**2. Import Library**

```python
from google_play_scraper import reviews_all
import pandas as pd
```

#**3. Scraping Data dari Google Play Store (Game Roblox)**

```python
# Scraping semua ulasan dari aplikasi Roblox di Play Store
# Package name Roblox: com.roblox.client
from google_play_scraper import Sort # Import Sort

app_id = 'com.roblox.client' # Define app_id

reviews = reviews_all(
    app_id,
    sleep_milliseconds=0,
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT
)

print(f"Jumlah review yang berhasil diambil: {len(reviews)}")
```

#**4. Ubah Hasil Scraping ke DataFrame**

```python
# Ubah hasil scraping menjadi DataFrame
df = pd.DataFrame(reviews)

# Tampilkan beberapa kolom penting
df = df[['userName', 'content', 'score', 'at']]
df.rename(columns={'userName':'nama_pengguna', 'content':'ulasan', 'score':'rating', 'at':'tanggal'}, inplace=True)

# Lihat contoh datanya
df.head()
```

#**5. Informasi dan Statistik Awal Dataset**

```python
print("Jumlah data:", len(df))
print(df['rating'].value_counts())

# Cek ada data kosong
print(df.isnull().sum())
```

#**6. Simpan Dataset ke File CSV**

```python
# Simpan data ke file CSV
df.to_csv('dataset_ulasan_roblox.csv', index=False, encoding='utf-8')
print("Dataset berhasil disimpan sebagai 'dataset_roblox.csv'")
```
